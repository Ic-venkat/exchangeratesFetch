```python
import requests
from datetime import datetime, timedelta
import pandas as pd

# Replace 'YOUR_ACCESS_KEY' with your actual API access key
access_key = 'API KEY Here'

start_date = datetime(2021, 1, 1).date()


# Define the target currencies (ZAR, EUR, USD)
target_currencies = 'ZAR,EUR,USD'

# Construct the URL for the API request

exchange_rate_data = pd.DataFrame(columns=['Date', 'ZAR', 'EUR', 'USD'])

def fetchExchangeRates(date):
    # Send an HTTP GET request to the API
    url = f'http://api.exchangeratesapi.io/v1/{date}?access_key={access_key}&symbols={target_currencies}'
    response = requests.get(url)
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        

        # Extract exchange rates for the target currencies
        zar_rate = data['rates']['ZAR']
        eur_rate = data['rates']['EUR']
        usd_rate = data['rates']['USD']
        exchange_rate_data.loc[len(exchange_rate_data)] = [date, zar_rate, eur_rate, usd_rate]

        # Print the exchange rates
        print(f'Exchange Rates on {date}:')
        print(f'{zar_rate} ZAR')
        print(f'{eur_rate} EUR')
        print(f'{usd_rate} USD')
        return data
```

```python
start_date = datetime(2023,10,4).date()
current_date = datetime.now().date()  # Adjust the current date to include the end date
while current_date >= start_date:
    formatted_date = start_date.strftime('%Y-%m-%d')
    fetchExchangeRates(formatted_date)
    start_date += timedelta(days=1)
```

```python
csv_filename = 'exchange_rates1.csv'
exchange_rate_data.to_csv(csv_filename, index=False)
```
